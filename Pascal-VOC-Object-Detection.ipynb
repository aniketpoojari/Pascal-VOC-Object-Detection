```python
import os
import cv2
import random
import numpy as np
import pandas as pd
from PIL import Image
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing

# for reading data
from detectron2.data import DatasetCatalog, MetadataCatalog
# loading pretrained weights
from detectron2 import model_zoo
# getting model config file
from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor, DefaultTrainer 
from detectron2.utils.visualizer import Visualizer
```

```python
dir = "drive/MyDrive/Pascal-VOC-Object-Detection/data/"
```

```python
# read files
train = pd.read_csv(dir + 'train.csv')
val = pd.read_csv(dir + 'val.csv')
```

```python
# Label encoding the classes
le = preprocessing.LabelEncoder().fit(train['class'])
train['class'] = le.transform(train['class'])
val['class'] = le.transform(val['class'])
```

```python
train.head()
```

```python
val.head()
```

```python
# define data loader funtion
def Train_Data_Converter(c = 0):
    count = 0
    dataset = []
    for img_name in train.filename.unique():

        det2_dict = {}
        img_dir = dir + 'train_images/' + img_name
        det2_dict['file_name'] = img_dir 
        det2_dict['width'], det2_dict['height'] = Image.open(img_dir).size
        det2_dict['image_id'] = img_name.split('.')[0]

        ## making annotations list
        annotations = []
        bbox_instances = train[train.filename == img_name][['xmin', 'ymin', 'xmax', 'ymax']].values.astype('float').tolist()
        category_instances = train[train.filename == img_name][['class']].values.astype('int').tolist()

        for idx in range(len(bbox_instances)):
            annot = {}
            annot['bbox'] = bbox_instances[idx]
            annot['bbox_mode'] = 0 
            annot['category_id'] = int(category_instances[idx][0])
            annotations.append(annot)

        det2_dict['annotations'] = annotations

        dataset.append(det2_dict)

        # Stop when number of images is satisfied
        count = count + 1
        if c != 0 and count == c:
          break
    return dataset
```

```python
# Adding custom dataset to the existing catalog
DatasetCatalog.register("train_data", Train_Data_Converter)
```

```python
# Defining related attributes of the dataset
Train_Data_Metadata = MetadataCatalog.get("train_data").set(thing_classes=list(le.classes_))
```

```python
for d in random.sample(Train_Data_Converter(2), 2):

    img = cv2.imread(d["file_name"])
    visualizer = Visualizer(img[:, :, ::-1], metadata = Train_Data_Metadata, scale = 0.5)
    vis = visualizer.draw_dataset_dict(d)
    
    plt.imshow(vis.get_image()[:, :, ::-1])
    plt.show()
```

```python
# define data loader funtion
def Val_Data_Converter(c = 0):
    count = 0
    dataset = []
    for img_name in val.filename.unique():

        det2_dict = {}
        img_dir = dir + 'val_images/' + img_name
        det2_dict['file_name'] = img_dir
        det2_dict['width'], det2_dict['height'] = Image.open(img_dir).size
        det2_dict['image_id'] = img_name.split('.')[0]

        # making annotations list
        annotations = []
        bbox_instances = val[val.filename == img_name][['xmin', 'ymin', 'xmax', 'ymax']].values.astype('float').tolist()
        category_instances = val[val.filename == img_name][['class']].values.astype('int').tolist()


        for idx in range(len(bbox_instances)):
            annot = {}
            annot['bbox'] = bbox_instances[idx]
            annot['bbox_mode'] = 0 
            annot['category_id'] = int(category_instances[idx][0])
            annotations.append(annot)

        det2_dict['annotations'] = annotations

        dataset.append(det2_dict)
        
        count = count + 1
        if c != 0 and count == c:
          break
    return dataset
```

```python
DatasetCatalog.register("val_data", Val_Data_Converter)
```

```python
Val_Data_Metadata = MetadataCatalog.get("val_data").set(thing_classes=list(le.classes_))
```

```python
for d in random.sample(Val_Data_Converter(2), 2):

    img = cv2.imread(d["file_name"])
    visualizer = Visualizer(img[:, :, ::-1], metadata = Val_Data_Metadata, scale = 0.5)
    vis = visualizer.draw_dataset_dict(d)
    
    plt.imshow(vis.get_image()[:, :, ::-1])
    plt.show()
```

## 4. Model Building

```python
## load default cfg file to define model
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/retinanet_R_50_FPN_1x.yaml")) 

cfg.DATASETS.TRAIN = ("train_data",) 
cfg.DATASETS.TEST = ("val_data",) 
cfg.DATALOADER.NUM_WORKERS = 2
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/retinanet_R_50_FPN_1x.yaml")  
cfg.SOLVER.IMS_PER_BATCH = 2 
cfg.SOLVER.BASE_LR = 0.00025  
cfg.SOLVER.MAX_ITER = 1000
cfg.MODEL.RETINANET.NUM_CLASSES = len(list(le.classes_)) 

os.makedirs(cfg.OUTPUT_DIR, exist_ok = True)
trainer = DefaultTrainer(cfg)        
trainer.resume_or_load(resume = True) 
trainer.train()
```

```python
cfg.MODEL.WEIGHTS = os.path.join(cfg.OUTPUT_DIR, "model_final.pth") 
cfg.MODEL.RETINANET.SCORE_THRESH_TEST = 0.5
cfg.DATASETS.TEST = ("test_data", ) 
predictor = DefaultPredictor(cfg) 
```

```python
## visualizing test data
img = Val_Data_Converter(4)[-1]
d = (Val_Data_Converter(4)[-1]).get('file_name')
im = cv2.imread(d)
v = Visualizer(im[:, :, ::-1], metadata = Val_Data_Metadata, scale = 0.5)
v = v.draw_dataset_dict(img)
cv2_imshow(v.get_image()[:, :, ::-1])
```
